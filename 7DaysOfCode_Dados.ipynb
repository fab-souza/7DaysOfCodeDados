Ou seja, identificar inconsistências, como campos que possuem valores nulos ou duplicados, converter campos, corrigir valores monetários, nomes incorretos, formatar campos de CNPJ, etc.

```python
import pandas as pd
```

```python
dataset2022 = pd.read_csv('/content/despesa_ceaps_2022.csv', sep=';', encoding='ISO-8859-1')
dataset2022
```

No arquivo original, a primeira linha informava a última atialização realizada. Isso fez com que o arquivo fosse interpretado com 12579 linhas e 2 colunas. Para ignorar a primeira linha do arquivo, utilizei o parâmetro ‘skiprow’.

```python
dataset2022 = pd.read_csv('/content/despesa_ceaps_2022.csv', sep=';', encoding='ISO-8859-1', skiprows=1)
dataset2022
```

```python
dataset2022.info()
```

Ao solicitar as informações do dataset, somente a coluna ‘DETALHAMENTO’ possui valores nulos. Dependendo do ‘TIPO_DESPESA’, esta falta de informação pode dificultar a análise do gasto.
Porém, os nulos podem ser preenchidos como ‘Não especificado’, ou algo similar, para não deixar lacunas no dataset.

```python
dataset2022['DETALHAMENTO'].fillna('Não informado', inplace=True)
```

Também há dados nulos na variável documento, que também serão substituídos por ‘Não informado’:

```python
dataset2022['DOCUMENTO'].fillna('Não informado', inplace=True)
```

```python
dataset2022.info()
```

Depois, a mudança de tipo de variável será para ‘DATA’:

```python
pd.to_datetime(dataset2022['DATA'].str.title(), format='%d/%m/%Y')
```

```python
dataset2022['DATA'] = pd.to_datetime(dataset2022['DATA'].str.title(), format='%d/%m/%Y')
dataset2022
```

Por último, retirar a vírgula nos valores reembolsados e transformar em float:

```python
dataset2022['VALOR_REEMBOLSADO'].str.replace(',','.').astype(float)
```

```python
dataset2022['VALOR_REEMBOLSADO'] = dataset2022['VALOR_REEMBOLSADO'].str.replace(',','.').astype(float)
dataset2022
```

Para finalizar o tratamento, verifiquei se há informações duplicadas:

```python
duplicados = dataset2022.duplicated()
duplicados.sum()
```

Com as variáveis tratadas, chegou a hora de adicionar os dados de anos anteriores (2018, 2019, 2020 e 2021):

```python
dados2018 = ('/content/despesa_ceaps_2018.csv')
dataset2018 = pd.read_csv(dados2018, sep=';', encoding='ISO-8859-1', skiprows=1)
dados2019 = ('/content/despesa_ceaps_2019.csv')
dataset2019 = pd.read_csv(dados2019, sep=';', encoding='ISO-8859-1', skiprows=1)
dados2020 = ('/content/despesa_ceaps_2020.csv')
dataset2020 = pd.read_csv(dados2020, sep=';', encoding='ISO-8859-1', skiprows=1)
dados2021 = ('/content/despesa_ceaps_2021.csv')
dataset2021 = pd.read_csv(dados2021, sep=';', encoding='ISO-8859-1', skiprows=1)
```

```python
from datetime import datetime, timezone
def ler_arquivo(nome_dataset): 
  nome_dataset['DETALHAMENTO'].fillna('Não informado', inplace=True)
  nome_dataset['DOCUMENTO'].fillna('Não informado', inplace=True)
  nome_dataset['VALOR_REEMBOLSADO'] = nome_dataset['VALOR_REEMBOLSADO'].str.replace(',','.').astype(float)
  nome_dataset['DATA'] = pd.to_datetime(nome_dataset['DATA'].str.title(), format='%d/%m/%Y', errors='coerce')
  nome_dataset['DATA'].replace(inplace = True)
```

```python
ler_arquivo(dataset2018)
```

```python
dataset2018.info()
```

```python
ler_arquivo(dataset2019)
```

```python
dataset2019.info()
```

```python
ler_arquivo(dataset2020)
```

```python
dataset2020.info()
```

```python
ler_arquivo(dataset2021)
```

```python
dataset2021.info()
```
